```python
from statsmodels.tsa.seasonal import seasonal_decompose

df_regresi['date'] = pd.to_datetime(df_regresi['date'])
df_regresi.set_index('date', inplace=True)
decomposed = seasonal_decompose(df_regresi)

plt.figure(figsize=(8, 8))

plt.subplot(311)
decomposed.trend.plot(ax=plt.gca())
plt.title('Trend')
plt.subplot(312)
decomposed.seasonal.plot(ax=plt.gca())
plt.title('Seasonality')
plt.subplot(313)
decomposed.resid.plot(ax=plt.gca())
plt.title('Residuals')

plt.tight_layout()
```

## Check Stationarity Data

```python
import statsmodels.api as sm

# Assuming your time series data is in a column named 'value'
ts = df_regresi['Qty']

# Perform the Augmented Dickey-Fuller test
result = sm.tsa.adfuller(ts)

# Extract the test statistic and p-value
adf_statistic, p_value, _, _, _, _ = result

# Print the results
print(f'ADF Statistic: {adf_statistic}')
print(f'p-value: {p_value}')

# Check if the time series is stationary based on the p-value
if p_value <= 0.05:  # You can adjust the significance level as needed
    print('The time series is stationary (reject the null hypothesis)')
else:
    print('The time series is non-stationary (fail to reject the null hypothesis)')

# Plot the time series for visual inspection
plt.plot(ts)
plt.title('Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

## Modelling

```python
def rmse(y_actual, y_pred):
    '''
    Fungsi untuk menghitung RMSE
    '''
    rmse_value = np.sqrt(mean_squared_error(y_actual, y_pred))
    return rmse_value

def eval(y_actual, y_pred):
    '''
    Fungsi untuk mengevaluasi model machine learning
    '''
    rmse_value = rmse(y_actual, y_pred)
    mae_value = mean_absolute_error(y_actual, y_pred)

    print(f'RMSE value: {rmse_value}')
    print(f'MAE value: {mae_value}')
```

```python
#ARIMA
df_train = df_train.set_index('date')
df_test = df_test.set_index('date')

y = df_train['Qty']

ARIMAmodel = ARIMA(y, order = (1, 0, 1))
ARIMAmodel = ARIMAmodel.fit()

y_pred = ARIMAmodel.get_forecast(len(df_test))

y_pred_df = y_pred.conf_int()
y_pred_df['predictions'] = ARIMAmodel.predict(start = y_pred_df.index[0], end = y_pred_df.index[-1])
y_pred_df.index = df_test.index
y_pred_out = y_pred_df['predictions']
eval(df_test['Qty'], y_pred_out)

plt.figure(figsize=(20,5))
plt.plot(df_train['Qty'])
plt.plot(df_test['Qty'], color='red')
plt.plot(y_pred_out, color='black', label='ARIMA Predictions')
plt.legend()
```

# Clustering

```python
 df_merge.head()
```

```python
# identifikasi dulu kolom-kolom yang redundant/korelasi tinggi
df_merge.corr()
```

```python
df_cluster = df_merge.groupby(['CustomerID']).agg({
    'TransactionID' : 'count',
    'Qty':'sum'
}).reset_index()
```

```python
df_cluster.head()
```

```python
data_cluster = df_cluster.drop(columns=['CustomerID'])

data_cluster_normalize = preprocessing.normalize(data_cluster)
```

```python
data_cluster_normalize
```

```python
K = range (2, 8)
fits = []
score = []

for k in K:
  model = KMeans(n_clusters = k, random_state=0, n_init='auto').fit(data_cluster_normalize)

  fits.append(model)

  score.append(silhouette_score(data_cluster_normalize, model.labels_, metric='euclidean'))
```

```python
# choose 4 cluster
sns.lineplot(x=K, y=score);
```

```python
fits[1]
```

```python
df_cluster['cluster_label'] = fits[1].labels_
```
